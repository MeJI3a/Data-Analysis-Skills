# 10. Normality Tests: 
#### **Introduction**
This project investigates whether the closing prices of stocks (ADANIPORTS dataset) follow a normal distribution. Normality tests are essential for validating assumptions in statistical modeling and hypothesis testing. The analysis includes visualizations and statistical tests to assess the distribution of closing prices.

#### **Objectives**
1. Visualize the distribution of closing prices using histograms and Q-Q plots.
2. Perform statistical normality tests:
   - Shapiro-Wilk Test
   - D'Agostino and Pearson's Test
3. Interpret the results to conclude whether the data is normally distributed.

---

```python
import pandas as pd
import itertools
import numpy as np
import plotly.graph_objects as go
df = pd.read_csv(r"C:\Users\Zana\Desktop\statistika\datasets\ADANIPORTS.csv")
df.head()
```

###  Visualizations for Normality Testing

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Close'], bins=30, kde=True)
plt.title('Histogram of Close Prices')
plt.xlabel('Close Price')
plt.ylabel('Frequency')
plt.show()
```

```python
import scipy.stats as stats

# Q-Q plot
plt.figure(figsize=(10, 6))
stats.probplot(df['Close'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Close Prices')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.grid()
plt.show()
```

### Perform Statistical Normality Tests

```python
from scipy.stats import shapiro, normaltest

# Shapiro-Wilk test
shapiro_stat, shapiro_p = shapiro(df['Close'])
print(f'Shapiro-Wilk Test: Statistic={shapiro_stat}, p-value={shapiro_p}')

# D'Agostino and Pearson's test
dagostino_stat, dagostino_p = normaltest(df['Close'])
print(f'D\'Agostino and Pearson\'s Test: Statistic={dagostino_stat}, p-value={dagostino_p}')
```

### **Findings**

#### **1. Visualizations**
- **Histogram**:
  - The histogram shows the frequency distribution of closing prices.
  - The curve suggests a roughly bell-shaped distribution, hinting at potential normality.
- **Q-Q Plot**:
  - The points largely deviate from the reference line, suggesting deviations from a perfect normal distribution.

#### **2. Statistical Tests**
- **Shapiro-Wilk Test**:
  - **Statistic**: `0.888`
  - **p-value**: `7.08e-44`
  - **Interpretation**: The p-value is much less than `0.05`, so we reject the null hypothesis. The data is not normally distributed.
- **D'Agostino and Pearson's Test**:
  - **Statistic**: `726.63`
  - **p-value**: `1.64e-158`
  - **Interpretation**: Again, the p-value is significantly less than `0.05`, indicating the data is not normally distributed.

---

### **Summary**
1. Visualizations suggest potential bell-shaped characteristics, but deviations from normality are evident in the Q-Q plot.
2. Both the Shapiro-Wilk and D'Agostino and Pearson's tests strongly reject the null hypothesis of normality.
3. The closing prices of the stock are not normally distributed.
