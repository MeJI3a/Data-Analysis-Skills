### Variance

```python
import pandas as pd

# Reading the Excel file
df = pd.read_excel(r"C:\Users\Zana\Desktop\statistika\Chapter01\1.2.xlsx")

# Calculate the mean value for the 'Order Values' column
mean_value = df['Order Values'].mean()

# Calculate the error (difference between the value and the mean)
df['Error'] = df['Order Values'] - mean_value

# Calculate the squared error
df['Squared Error'] = df['Error'] ** 2

# Calculate the sum of squared errors
sum_squared_error = df['Squared Error'].sum()

# Number of elements in the sample (n)
n = len(df)

# Calculate the variance as the sum of squared errors divided by n
variance = sum_squared_error / n

# Output the variance
print(f"Variance: {variance}")
```

### Standard Deviation (long way)

```python
import math

# Use the variance calculated earlier
variance = sum_squared_error / n  # Or for sample variance: variance = sum_squared_error / (n - 1)

# Calculate the standard deviation
std_deviation = math.sqrt(variance)

# Output the standard deviation
print(f"Standard Deviation: {std_deviation}")
```

### Sample variance | Sample standard deviation

```python
# Calculating the mean of the 'Order Values' column
mean_value = df['Order Values'].mean()

# Calculating the error (difference between each value and the mean)
df['Error'] = df['Order Values'] - mean_value

# Calculating the squared error
df['Squared Error'] = df['Error'] ** 2

# Number of elements (n)
n = len(df)

# Calculating sample variance (VAR.S)
var_s = df['Squared Error'].sum() / (n - 1)

# Calculating sample standard deviation (STDEV.S)
stdev_s = math.sqrt(var_s)

# Displaying the results
print(f"Sample variance (VAR.S): {var_s}")
print(f"Sample standard deviation (STDEV.S): {stdev_s}")
```

### Central Limit Theorem

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Reading the Excel file
df = pd.read_excel(r"C:\Users\Zana\Desktop\statistika\Chapter01\1.2.xlsx")

# Extracting the 'Order Values' column
order_values = df['Order Values']

# Number of samples we want to draw and size of each sample
n_samples = 1000  # Number of samples
sample_size = 30  # Size of each sample

# List to store means of each sample
sample_means = []

# Generating n_samples of sample_size from the 'Order Values' column
for _ in range(n_samples):
    sample = np.random.choice(order_values, size=sample_size, replace=True)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)

# Plotting the distribution of sample means
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='g')
plt.title('Distribution of Sample Means 30 (Central Limit Theorem)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```

### Three Stigma Rule

```python
import pandas as pd

# Reading the Excel file
df = pd.read_excel(r"C:\Users\Zana\Desktop\statistika\Chapter01\1.2.xlsx")

# Calculate the mean and standard deviation
mean_value = df['Order Values'].mean()
std_deviation = df['Order Values'].std()

# Calculate ranges for 68%, 95%, and 99.7% using the empirical rule (three-sigma rule)
range_68 = (mean_value - std_deviation, mean_value + std_deviation)
range_95 = (mean_value - 2 * std_deviation, mean_value + 2 * std_deviation)
range_997 = (mean_value - 3 * std_deviation, mean_value + 3 * std_deviation)

# Output the results
print(f"Mean value: {mean_value}")
print(f"Standard deviation: {std_deviation}")
print(f"68% of data falls within the range: {range_68}")
print(f"95% of data falls within the range: {range_95}")
print(f"99.7% of data falls within the range: {range_997}")
```

### Margin of Error

```python
import math
import pandas as pd

# Reading the Excel file
df = pd.read_excel(r"C:\Users\Zana\Desktop\statistika\Chapter01\1.2.xlsx")

# Calculate the sample standard deviation
std_deviation = df['Order Values'].std()

# Sample size
n = len(df)

# Z-value for a 95% confidence interval
Z = 2.58

# Calculate the margin of error
margin_of_error = Z * (std_deviation / math.sqrt(n))

# Output the result
print(f"Margin of Error: {margin_of_error}")
```

```python
import math
import pandas as pd

# Reading the Excel file
df = pd.read_excel(r"C:\Users\Zana\Desktop\statistika\Chapter01\1.2.xlsx")

# Calculate the sample standard deviation
std_deviation = df['Order Values'].std()

# Sample size
n = len(df)

# Z-values for various confidence levels
confidence_levels = {
    "80%": 1.28,
    "90%": 1.645,
    "95%": 1.96,
    "98%": 2.33,
    "99%": 2.58
}

print(df)

# Calculate margin of error for each confidence level
for confidence, Z in confidence_levels.items():
    margin_of_error = Z * (std_deviation / math.sqrt(n))
    print(f"Margin of Error for {confidence} confidence level: {margin_of_error}")
```
